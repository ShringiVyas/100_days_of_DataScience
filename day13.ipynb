Poisson Distribution

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


#load the dataset
iris = load_iris()


#select sepal length data for iris setosa
setosa_sepal_length = iris.data[iris.target == 0, 0]

#calculating lambda parameter for poisson distribution (mean of sepal length)
lambda_param = np.mean(setosa_sepal_length)

#poisson distribution calcualtion
poisson_dist = poisson(mu=lambda_param)

#generate x values (sepal length)
x_values = np.arange(0, 10, 0.1)

#plot poisson distribution
plt.plot(x_values, poisson_dist.pmf(x_values))
plt.xlabel('sepal length')
plt.ylabel('probability mass')
```

Normal Distribution

```python
import numpy as np
from sklearn.datasets import load_iris
from scipy.stats import norm
import  matplotlib.pyplot as plt

#load iris dataset
iris = load_iris()

#select sepal length data for all iris flowers
sepal_length_all = iris.data[:, 0]

#calculating mean and standard deviation of sepal lengths
mean_sepal_length = np.mean(sepal_length_all)
std_deviation_sepal = np.std(sepal_length_all)

#normal distribution calculation
normal_dist = norm(loc=mean_sepal_length, scale=std_deviation_sepal)

#ploting normal distribution
plt.plot(x_values, normal_dist.pdf(x_values), label=f'Normal Distribution (μ={mean_sepal_length:.2f}, σ={std_deviation_sepal:.2f})')
plt.xlabel('sepal length')
plt.ylabel('probability density')
```
